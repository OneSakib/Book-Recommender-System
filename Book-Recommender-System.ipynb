```python
ratings=ratings[['User-ID','ISBN','Book-Rating']]
```

```python
x=ratings['user_id'].value_counts()>200
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
rating_with_books=ratings.merge(books,on='isbn')
```

```python
rating_with_books.head()
```

```python
rating_with_books.shape
```

```python
number_rating=rating_with_books.groupby('title')['rating'].count().reset_index().rename(columns={'rating':'number of rating'})
```

```python
number_rating.head()
```

```python
final_rating=rating_with_books.merge(number_rating,on='title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['number of rating']>=50]
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot.head(1)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
pd.DataFrame.sparse.from_spmatrix(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in suggestions:
    print(book_pivot.index[i])
```

```python
np.where(book_pivot.index=='Animal Farm')[0][0]
```

```python
def recommanded_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0]
     if book_id.size == 0:
        print("Book Not Found")
        return
    book_id=book_id[0]
    distances,suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)    
    for i in suggestions[1:]:
        print(book_pivot.index[i])
    
        
```

```python
recommanded_book('Animal Farm')
```
